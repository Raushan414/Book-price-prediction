```python
def clean_text(text):
    text = text.lower()
    text = re.sub(r'@[a-zA-Z0-9_]+', '', text)   
    text = re.sub(r'https?://[A-Za-z0-9./]+', '', text)   
    text = re.sub(r'www.[^ ]+', '', text)  
    text = re.sub(r'[a-zA-Z0-9]*www[a-zA-Z0-9]*com[a-zA-Z0-9]*', '', text)  
    text = re.sub(r'[^a-zA-Z]', ' ', text)   
    text = [token for token in text.split() if len(token) > 2]
    text = ' '.join(text)
    return text

train_df1['Synopsis'] = train_df1['Synopsis'].apply(clean_text)
test_df1['Synopsis'] = test_df1['Synopsis'].apply(clean_text)

train_df1['Title'] = train_df1['Title'].apply(clean_text)
test_df1['Title'] = test_df1['Title'].apply(clean_text)
```

```python
def evaluate(y_pred,y_real):
    mse=metrics.mean_squared_error(y_real, y_pred)
    loss = np.sqrt(mse)
    return loss
```

```python
target=train_df1['Price']
train_data=train_df1.drop(['Price'],axis=1)
target=np.array(target)
train_data=np.array(train_data)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import SVR
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeRegressor
```

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
ss.fit(train_data)
train_data=ss.transform(train_data)
```

```python
X_train,X_val,y_train,y_val=train_test_split(train_data,target,test_size=0.2,shuffle=True)
```

```python
linreg = LinearRegression()
linreg.fit(X_train,y_train)
train_sc=evaluate(linreg.predict(X_train),y_train)
val_sc=evaluate(linreg.predict(X_val),y_val)
print("Linear Regression: Training loss-{} Val loss-{}".format(train_sc,val_sc))

knn = KNeighborsRegressor()
knn.fit(X_train,y_train)
train_sc=evaluate(knn.predict(X_train),y_train)
val_sc=evaluate(knn.predict(X_val),y_val)
print("knn: Training loss-{} Val loss-{}".format(train_sc,val_sc))

dtr = DecisionTreeRegressor()
dtr.fit(X_train,y_train)
train_sc=evaluate(dtr.predict(X_train),y_train)
val_sc=evaluate(dtr.predict(X_val),y_val)
print("Decision Tree: Training loss-{} Val loss-{}".format(train_sc,val_sc))

svm = SVR()
svm.fit(X_train,y_train)
train_sc=evaluate(svm.predict(X_train),y_train)
val_sc=evaluate(svm.predict(X_val),y_val)
print("Support Vecor Machine: Training loss-{} Val loss-{}".format(train_sc,val_sc))
```
